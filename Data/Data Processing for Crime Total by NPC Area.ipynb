```python
import pandas as pd
```

Read in crime total data and npc area size data

```python
crime_total = pd.read_csv('Total Crime.csv')
npc_area = pd.read_csv('NPC Area Size.csv')
```

```python
crime_total
```

```python
npc_area
```

Group by the NPC Name and sum their area

```python
unique_npc = npc_area.groupby(["NPC_NAME"]).sum().reset_index()
unique_npc
```

```python
type(unique_npc)
```

Drop Airport division row

```python
unique_npc = unique_npc.drop(0)
unique_npc
```

```python
total_crime_and_area = unique_npc.merge(crime_total, left_on = "NPC_NAME", right_on = "npc")
total_crime_and_area
```

```python
total_crime_and_area.info()
```

```python
#convert m2 to km2
total_crime_and_area['NPC Area in km2'] = total_crime_and_area['Area of NPC']/1000000
total_crime_and_area
```

```python
total_crime_and_area['crime rate npc area'] = total_crime_and_area['crime_count']/total_crime_and_area['NPC Area in km2']
```

```python
total_crime_and_area
```

```python
total_crime_and_area = total_crime_and_area[['NPC_NAME', 'Area of NPC', 'crime_count', 'NPC Area in km2', 'crime rate npc area']]
total_crime_and_area
```

```python
total_crime_and_area.to_csv("Data/Crime Rate in Each NPC.csv")
```

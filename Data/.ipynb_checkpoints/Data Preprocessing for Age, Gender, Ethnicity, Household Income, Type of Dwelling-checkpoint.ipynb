```python
import pandas as pd
```

```python
age = pd.read_excel("AgeGroup_2020_Cleaned.xlsx")
gender = pd.read_excel("Gender_2020_Cleaned.xlsx")
ethnicity = pd.read_csv("Ethnicity Cleaned.csv")
income = pd.read_excel("Household_Income_Cleaned.xlsx")
dwelling = pd.read_csv("Type of Dwelling cleaned.csv")
```

```python
ethnicity = ethnicity.dropna()
ethnicity.reset_index(inplace=True)
ethnicity = ethnicity.drop(columns=["index"])
```

```python
age
```

```python
for i in range(len(age)):
    age.loc[i, "Planning Area/Subzone of Residence"] = age.loc[i, "Planning Area/Subzone of Residence"].strip()
```

```python
for i in range(len(gender)):
    gender.loc[i, "Planning Area/Subzone of Residence"] = gender.loc[i, "Planning Area/Subzone of Residence"].strip()
```

```python
for i in range(len(ethnicity)):
    ethnicity.loc[i, "Planning Area/Subzone of Residence"] = ethnicity.loc[i, "Planning Area/Subzone of Residence"].strip()
```

```python
area_list = list(income["Planning Area of Residence"]) 
gender_index = []
count = 0
for i in range(len(gender)):
    if gender.loc[i, "Planning Area/Subzone of Residence"] in area_list:
        gender_index.append(i)
        count += 1
count
```

```python
gender_cleaned = gender.loc[gender_index]
gender_cleaned.reset_index(inplace=True)
gender_cleaned = gender_cleaned.drop(columns=["index"])
gender_cleaned
```

```python
gender_cleaned.to_excel("Gender_Cleaned_Subset.xlsx", index = False)
```

```python
area_list = list(income["Planning Area of Residence"]) 
age_index = []
count = 0
for i in range(len(age)):
    if age.loc[i, "Planning Area/Subzone of Residence"] in area_list:
        age_index.append(i)
        count += 1
count
```

```python
age_cleaned = age.loc[age_index]
age_cleaned.reset_index(inplace=True)
age_cleaned = age_cleaned.drop(columns=["index"])
age_cleaned
```

```python
age_cleaned.to_excel("Age_Cleaned_Subset.xlsx", index = False)
```
